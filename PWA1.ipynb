Mutable objects: Mutable means that can be modified or changed.In Python,there are such kind of objects that allow to modify once they have been created.Objects of list,set,dictionaries are mutables.
Example:


```
list=[1,2,3,4,5,6]
list[0]=8 # This is allowable.
s1={23,45,67,89}
s1[3]=56 # This is also allowable
dic={"a":1,"b":2,"c":3}
dic["d"]=dic["b"]
del dic["b"]
```


Immutable objects: In contrast of mutable objects,immutable objects do not allow to modify once  they have been created.Objects of string and tuple are immutables.
Example:

```
str="Python Assignment"
str[2]="c" # This is not allowed
tup=(1,2,3,4)
tup[0]=5 # This is also not allowed
```
